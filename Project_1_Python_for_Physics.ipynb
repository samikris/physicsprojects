Creating Functions for Position/Velocity

```python
def xpos(x0, t, vx0, ax):
  x = x0 + vx0*t + ((ax)/2)*(t**2)
  return x
```

```python
def ypos(y0, t, vy0, ay):
  y = y0 + vy0*t + ((ay)/2)*(t**2)
  return y
```

```python
def xvel(vx0, ax, t):
  vx = vx0 + (ax)*t
  return vx
```

```python
def yvel(vy0, ay, t):
  vy = vy0 + (ay)*t
  return vy
```

Plotting the functions

```python
#creating t values
t = np.linspace(0, 5, 100)
print(t)
```

```python
#example values
x0 = 0
vx0 = 20
ax = 0
```

List comprehension format to get position for every t value

```python
x_values = [xpos(x0, time, vx0, ax) for time in t] #list comprehension --> gimme another array by applying another function to every element
print(x_values)
```

X Position vs Time

```python
plt.subplot(311)
plt.plot(t,x_values)
plt.xlabel("Time (s)")
plt.ylabel("Horizontal Pos (m)")
plt.title('Projectile Position X vs. Time')
```

Y Position vs Time

```python
y0 = 0
vy0 = 50
ay = -9.8
y_values = [ypos(y0, time, vy0, ay) for time in t] #list comprehension --> gimme another array by applying another function to every element
plt.subplot(311)
plt.plot(t,y_values)
plt.xlabel("Time (s)")
plt.ylabel("Vertical Pos (m)")
plt.title('Projectile Position Y vs. Time')
```

Vx vs Time

```python
vx0 = 20
ax = 0
vx_values = [xvel(vx0, ax, time) for time in t]
plt.subplot(311)
plt.plot(t,vx_values)
plt.xlabel("Time (s)")
plt.ylabel("Horizontal Velocity (m)")
plt.title('X Velocity vs. Time')
```

Vy vs Time

```python
vy_values = [yvel(vy0, ay, time) for time in t]
plt.subplot(311)
plt.plot(t,vy_values)
plt.xlabel("Time (s)")
plt.ylabel("Horizontal Velocity (m)")
plt.title('Y Velocity vs. Time')

#print(vy_values)
```

X vs Y Position Graph

```python
plt.subplot(311)
plt.plot(x_values,y_values)
plt.xlabel("Horizontal Pos (m)")
plt.ylabel("Vertical Pos (m)")
plt.title('X vs Y Position')
```
